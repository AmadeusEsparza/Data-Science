````python
# Importar librerías
import numpy as np
import matplotlib.pyplot as plt
````

````python
# Definir variables globales
x = np.linspace(-10, 10, 1000)
x
````

````python
# Definir función de derivación
def derivada(y):
  return np.diff(y)/np.diff(x)
````

````python
# Definir función de antiderivación
def antiderivada(y):
  return np.cumsum(((y[:-1] + y[1:]) * np.diff(x))/2)
````

````python
# Definir función de visualización
def visualizacion(y, dy, sy):
  plt.plot(x, y, color = 'blue', label = 'función')
  plt.plot(x[1:], dy, color = 'green', label = 'derivada')
  plt.plot(x[1:], sy, color = 'purple', label = 'antiderivada')
  plt.axhline(y = 0, color = 'red')
  plt.axvline(x = 0, color = 'red')
  plt.legend(loc = 'upper left')
  plt.grid()
  plt.show()
````

## **FUNCIÓN LINEAL**

Una función lineal tiene la forma:

\begin{equation}
y = mx + b
\end{equation}

Programar la siguiente función:

\begin{equation}
f(x) = x
\, \, \, \, \, \, \, \, \, \, \, \,
\Rightarrow\,
\, \, \, \, \, \, \, \, \, \, \, \,
f'(x) = 1
\, \, \, \, \, \, \, \, \, \, \, \,
\wedge
\, \, \, \, \, \, \, \, \, \, \, \,
F(x) = \frac{x^2}{2}
\end{equation}

````python
# Definir función
def funcion_lineal(x):
  return x

f_lineal = funcion_lineal(x)
dy_lineal = derivada(f_lineal)
sy_lineal = antiderivada(f_lineal) - np.median(antiderivada(f_lineal))

````

````python
np.median(sy_lineal - np.median(sy_lineal))
````

````python
# Usar función de visualización
visualizacion(f_lineal, dy_lineal, sy_lineal)
````

## **FUNCIÓN CÚBICA**

Una función cúbica tiene la forma:

\begin{equation}

\end{equation}

Programar la siguiente función:

\begin{equation}

\, \, \, \, \, \, \, \, \, \, \, \,
\Rightarrow

\, \, \, \, \, \, \, \, \, \, \, \,
\wedge

F(x) = \frac{x^4}{4}

````python
# Definir función

def funcion_cubica(x):
  return np.power(x, 3)

f_cubica = funcion_cubica(x)
dy_cubica = derivada(f_cubica)
sy_cubica = antiderivada(f_cubica) - np.median(antiderivada(f_cubica))
````

````python
# Usar función de visualización
visualizacion(f_cubica, dy_cubica, sy_cubica)
````
